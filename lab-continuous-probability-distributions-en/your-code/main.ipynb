## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
def generate_uniform(bottom, ceiling, count):
    """
    Generates uniformly distributed random numbers
    
    Parameters:
    bottom - the lower boundary of the generated numbers
    ceiling - the upper boundary of the generated numbers
    count - how many numbers to generate
    
    Returns:
    array of uniformly distributed random numbers
    """
    return np.random.uniform(bottom, ceiling, count)
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
def generate_exponential(mean, size):
    """
    Generates exponentially distributed random numbers
    
    Parameters:
    mean - the mean of the distribution (scale = mean)
    size - how many numbers to generate
    
    Returns:
    array of exponentially distributed random numbers
    """
    # For exponential, scale = mean
    return np.random.exponential(scale=mean, size=size)
```

How are the two distributions different?

```python
# Generate with size 10
exp_10 = generate_exponential(10, 10)

# Generate with size 100
exp_100 = generate_exponential(10, 100)
```

The mean changes, so the distribution changes as well. 

```python
# Create a figure with 2 subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot first exponential distribution (size=10)
ax1.hist(exp_10, bins=10, edgecolor='black', color='gold')
ax1.set_title('Exponential Distribution: Mean=10, Size=10')
ax1.set_xlabel('Value')
ax1.set_ylabel('Count')

# Plot second exponential distribution (size=100)
ax2.hist(exp_100, bins=100, edgecolor='black', color='darkorange')
ax2.set_title('Exponential Distribution: Mean=10, Size=100')
ax2.set_xlabel('Value')
ax2.set_ylabel('Count')

plt.tight_layout()
plt.show()
```

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# For exponential distribution with mean = 10 minutes
# lambda = 1/mean = 0.1
# P(x < 15) = 1 - e^(-lambda * 15)

mean = 10
lambda_param = 1 / mean

# Calculate probability of less than 15 minutes
prob_less_than_15 = 1 - np.exp(-lambda_param * 15)

print(f"Probability of spending less than 15 minutes: {prob_less_than_15:.4f}")
print(f"This is {prob_less_than_15 * 100:.2f}%")
```

What is the probability that the customer will spend more than 15 minutes

```python
# P(x > 15) = 1 - P(x < 15)
# Or directly: P(x > 15) = e^(-lambda * 15)

prob_more_than_15 = np.exp(-lambda_param * 15)

print(f"Probability of spending more than 15 minutes: {prob_more_than_15:.4f}")
print(f"This is {prob_more_than_15 * 100:.2f}%")

# Alternative calculation using 1 - P(less than 15)
print(f"\nCheck: 1 - {prob_less_than_15:.4f} = {1 - prob_less_than_15:.4f}")
```
